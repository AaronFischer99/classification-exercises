```python
import pandas as pd
from sklearn.model_selection import train_test_split

def split_data(df, target_column, test_size=0.2, validate_size=0.1, random_state=42):
# def split_data(X, y, test_size=0.2, validate_size=0.1, random_state='')
    """
    Split the data into train, test, and validate datasets.

    Parameters:
    - df: DataFrame, input data
    - target_column: str, name of the target column
    - test_size: float, proportion of the data to include in the test split
    - validate_size: float, proportion of the data to include in the validation split
    - random_state: int, seed for random number generation

    Returns:
    - train, test, validate DataFrames
    """
    train, test = train_test_split(df, test_size=test_size, random_state=random_state)
    train, validate = train_test_split(train, test_size=validate_size/(1-test_size), random_state=random_state)

    return train, test, validate

if __name__ == "__main__":
    # Example usage
    data = pd.read_csv("iris_df.csv")  # Replace with your actual data file
    train_data, test_data, validate_data = split_data(data, target_column="your_target_column")

    # You can now use train_data, test_data, and validate_data in your machine learning pipeline
```

from sklearn.model_selection import train_test_split



def split_data(df):
    '''
    take in a DataFrame and return train, validate, and test DataFrames; stratify on survived.
    return train, validate, test DataFrames.
    '''
    #test size
    train_validate, test = train_test_split(df, test_size=.2, random_state=123, stratify=df.survived)
    
    #validate size
    train, validate = train_test_split(train_validate, 
                                       test_size=.3, 
                                       random_state=123, 
                                       stratify=train_validate.survived)
    return train, validate, test


```python
import pandas as pd
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target

# Split the Iris dataset
train_iris, test_iris, validate_iris = split_data(iris_df, target_column="target")

# Display the shapes of the resulting datasets
print("Train Iris Shape:", train_iris.shape)
print("Test Iris Shape:", test_iris.shape)
print("Validate Iris Shape:", validate_iris.shape)
```
